```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
#import pandas_profiling as ppf  #eda
from sklearn.preprocessing import LabelEncoder #标签编码
from sklearn.preprocessing import MinMaxScaler #归一化
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split #数据集的划分
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_absolute_error #评估函数
from sklearn import metrics
%matplotlib inline
```

## 数据处理

```python
def read_and_process(file):
    data = pd.read_csv(file)
    # 缺失值处理
    data['Age'] = data['Age'].fillna(np.mean(data['Age']))
    data['Fare'] = data['Fare'].fillna(np.mean(data['Fare']))
    
    #将字符串做成onehot数值型
    data_dummy = pd.get_dummies(data[['Sex', 'Embarked']])
    data = data.drop(labels=['Embarked', 'Sex'], axis=1)
    data = data.join(data_dummy)
    
    #删除没必要的列
    data = data.drop(labels=['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1)
    
    #数据归一化
    stdsc = StandardScaler()
    data['Age'] = stdsc.fit_transform(np.array(data['Age']).reshape(-1, 1))
    data['Fare'] = stdsc.fit_transform(np.array(data['Fare']).reshape(-1, 1))
    
    return data
```

```python
test = pd.read_csv('input/test.csv')
test_PassengerId = test['PassengerId']
```

```python
train = read_and_process('input/train.csv')
test = read_and_process('input/test.csv')
```

## 数据建模
### 1.划分数据集

```python
x = train.drop('Survived', axis=1)
y = train['Survived']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape, y_train.shape
```

### 2.搭建模型，训练

```python
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
print('train score ', clf.score(x_train, y_train)) #过拟合
print('test score ', clf.score(x_test, y_test))
```

### 3.优化模型
1. 通过设置max_depth来限制模型深度


```python
def clf_score(depth):
    clf = DecisionTreeClassifier(max_depth=depth)
    clf.fit(x_train, y_train)
    return (clf.score(x_train, y_train), clf.score(x_test, y_test))
```

```python
depth = np.arange(1, 10)
scores = [clf_score(d) for d in depth]
tr_scores = [s[0] for s in scores]
te_scores = [s[1] for s in scores]

tr_best_index = np.argmax(tr_scores)
te_best_index = np.argmax(te_scores)

print('best depth {}'.format(depth[te_best_index]))
print('best score {}'.format(te_scores[te_best_index]))
```

```python
plt.figure(figsize=(6, 4), dpi=120)
plt.grid()
plt.xlabel('max depth of decision tree')
plt.ylabel('Scores')
plt.plot(depth, te_scores, label='test_scores')
plt.plot(depth, tr_scores, label='train_scores')
plt.legend()
```

2. 通过min_impurity_decrease来优化模型
- 指定基尼不纯度的阙值，当决策树分裂后，其信息增益低于这个阙值时不再分裂

```python
def clf_score(val):
    clf = DecisionTreeClassifier(criterion='gini', min_impurity_decrease=val)
    clf.fit(x_train, y_train)
    return (clf.score(x_train, y_train), clf.score(x_test, y_test))
```

```python
vals = np.linspace(0, 0.005, 1000)
scores = [clf_score(v) for v in vals]
tr_scores = [s[0] for s in scores]
te_scores = [s[1] for s in scores]

tr_best_index = np.argmax(tr_scores)
te_best_index = np.argmax(te_scores)

print('best val {}'.format(vals[te_best_index]))
print('best score {}'.format(te_scores[te_best_index]))
```

```python
plt.figure(figsize=(6, 4), dpi=120)
plt.grid()
plt.xlabel('max depth of decision tree')
plt.ylabel('Scores')
plt.plot(vals, te_scores, label='test_scores')
plt.plot(vals, tr_scores, label='train_scores')
plt.legend()
```

模型评估

```python
model = DecisionTreeClassifier(max_depth=6, min_impurity_decrease=0.00075)
model.fit(x_train, y_train)
print('tree score {}'.format(model.score(x_test, y_test)))

y_pred = model.predict(x_test)

print('查准率：', metrics.precision_score(y_test, y_pred))
print('查全率：', metrics.recall_score(y_test, y_pred))
print('F1_score: ', metrics.f1_score(y_test, y_pred))
```

3. 使用GridSearchCV求最优参数

```python
from sklearn.model_selection import GridSearchCV

thresholds = np.linspace(0, 0.1, 100)
param_grid = {'min_impurity_decrease':thresholds, 'max_depth': np.arange(1, 10)}

clf = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5) #cv交叉验证
clf.fit(x, y)

print('best_parameters:{0}\n best_score:{1}'.format(clf.best_params_, clf.best_score_))
```

关键参数cv，用来指定交叉验证数据集的生成规则。cv=5表示每次计算把数据集分成5份，拿其中一份作为交叉验证数据集，其他座位训练集

cvl.cv_results_里面保存了计算过程中的所有中间结果

### 3.提交

```python
predict = clf.predict(test)
```

```python
submission = pd.DataFrame({'PassengerId':test_PassengerId, 'Survived': predict})
```

```python
submission.to_csv('submission.csv', index=False)
```
