```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize

def plot_data_internal(X, y):
    x_min, x_max = X[ :, 0 ].min() - .5, X[ :, 0 ].max() + .5
    y_min, y_max = X[ :, 1 ].min() - .5, X[ :, 1 ].max() + .5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    plt.figure()
    plt.xlim(xx.min(None), xx.max(None))
    plt.ylim(yy.min(None), yy.max(None))
    ax = plt.gca()
    ax.plot(X[y == 0, 0], X[y == 0, 1], 'ro', label = 'Class 1')
    ax.plot(X[y == 1, 0], X[y == 1, 1], 'bo', label = 'Class 2')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Plot data')
    plt.legend(loc = 'upper left', scatterpoints = 1, numpoints = 1)
    return xx, yy

def plot_data(X, y):
    xx, yy = plot_data_internal(X, y)
    plt.show()
    
def plot_ll(ll):
    plt.figure()
    ax = plt.gca()
    #plt.xlim(0, len(ll) + 2)
    plt.ylim(min(ll) - 0.1, max(ll) + 0.1)
    ax.plot(np.arange(1, len(ll) + 1), ll, 'r-')
    ax.set_xscale('log')
    plt.xlabel('Steps')
    plt.ylabel('Average log-likelihood')
    plt.title('Plot Average Log-likelihood Curve')
    plt.show()

def plot_ll2(ll1, ll2):
    plt.figure()
    ax = plt.gca()
    y_min = min(min(ll1), min(ll2))
    y_max = max(max(ll1),max(ll2))
    
    plt.ylim(y_min - 0.1, y_max + 0.1)
    ax.plot(np.arange(1, len(ll1) + 1), ll1, 'r-', label="train")
    ax.plot(np.arange(1, len(ll2) + 1), ll2, 'g-', label="test")
    ax.set_xscale('log')
    plt.xlabel('Steps')
    plt.ylabel('Average log-likelihood')
    plt.title('Plot Average Log-likelihood Curve, n=5e-2')
    plt.legend()
    plt.show()
```

```python
def plot_predictive_distribution(X, y, w, map_inputs = lambda x : x):
    xx, yy = plot_data_internal(X, y)
    ax = plt.gca()
    X_tilde = get_x_tilde(map_inputs(np.concatenate((xx.ravel().reshape((-1, 1)), yy.ravel().reshape((-1, 1))), 1)))
    Z = predict(X_tilde, w)
    Z = Z.reshape(xx.shape)
    cs2 = ax.contour(xx, yy, Z, cmap = 'RdBu', linewidths = 2)
    plt.clabel(cs2, fmt = '%2.1f', colors = 'k', fontsize = 14)
    plt.show()

def evaluate_basis_functions(l, X, Z):
    X2 = np.sum(X**2, 1)
    Z2 = np.sum(Z**2, 1)
    ones_Z = np.ones(Z.shape[ 0 ])
    ones_X = np.ones(X.shape[ 0 ])
    r2 = np.outer(X2, ones_Z) - 2 * np.dot(X, Z.T) + np.outer(ones_X, Z2)
    return np.exp(-0.5 / l**2 * r2)
```

```python
X = np.loadtxt('X.txt')
y = np.loadtxt('y.txt')

# We randomly permute the data

permutation = np.random.permutation(X.shape[ 0 ])
X = X[ permutation, : ]
y = y[ permutation ]

def preprocess(x):
    mean = np.mean(x)
    std = np.std(x)
    return (x - mean)/std, mean, std
```

```python
def logistic(x):
    return 1.0 / (1.0 + np.exp(-x))

def predict(X_tilde, w): return logistic(np.dot(X_tilde, w))

def compute_average_ll(X_tilde, y, w):
    output_prob = predict(X_tilde, w)
    return np.mean(y * np.log(output_prob) + (1 - y) * np.log(1.0 - output_prob))

def get_x_tilde(X): return np.concatenate((np.ones((X.shape[ 0 ], 1 )), X), 1)

def fit_w(X_tilde_train, y_train, X_tilde_test, y_test, n_steps, alpha):
    w = np.random.randn(X_tilde_train.shape[ 1 ])
    ll_train = np.zeros(n_steps)
    ll_test = np.zeros(n_steps)
    for i in range(n_steps):
        sigmoid_value = predict(X_tilde_train, w)
        w = w + alpha * np.dot(X_tilde_train.T, y_train - sigmoid_value)
        ll_train[ i ] = compute_average_ll(X_tilde_train, y_train, w)
        ll_test[ i ] = compute_average_ll(X_tilde_test, y_test, w)
    return w, ll_train, ll_test

def log_likelihood(X, y, w):
    sigmoid_value = logistic(np.dot(X, w))
    y2 = 1 - y
    loglikelihood = np.dot(y.T, np.log(sigmoid_value)) + np.dot(y2.T, np.log(1 - sigmoid_value))
    return loglikelihood

def posterior_fprime(w, *args):
    X = args[0]
    y = args[1]; y2 = 1 - y
    variance = args[2]
    
    sigmoid_value = logistic(np.dot(X, w))
    penalty = np.dot(w.T, w) / (2*variance)
    func = log_likelihood(X, y, w) - penalty
    fprime = np.dot(X.T, y - sigmoid_value) - w / variance
    
    # note the scipy optimization only return the minimum value
    # however argmax_func needed, hence use negative value
    return -1*func, -1*fprime

def obtain_map(X, w, y, variance):
    optm_result = scipy.optimize.fmin_l_bfgs_b(posterior_fprime, x0=w, args=(X, y, variance), factr=10)
    return optm_result

# def test_obtain_map():
#     w = np.random.randn(X_tilde_train.shape[ 1 ])
#     result = obtain_map(X_tilde_train, w, y_train, variance=1)
#     wmap = result[0]
#     print('wmap[:10]: ', wmap[:10])
#     print('warnflag: ', result[2]['warnflag'])
#     print('funcalls:' ,result[2]['funcalls'])
#     print('grad_sum: ', result[2]['grad'].sum())
#     return None

def bayesian_training(X_tilde_train, y_train, X_tilde_test, y_test, variance):
    N_example = X_tilde_train.shape[0]; D_feature = X_tilde_train.shape[1]

    #obtain wmap solution
    w = np.random.randn(D_feature)
    result = obtain_map(X_tilde_train, w, y_train, variance)
    wmap = result[0] #optm_result[0] is the optimised value

    #calculate S_N matrix
    S_0_inv = np.zeros((D_feature, D_feature)) # shape: D*D
    np.fill_diagonal(S_0_inv, 1/variance)
    y0 = logistic(np.dot(X_tilde_train, wmap)).reshape((N_example,1)) #shape N*1
    alpha = y0 * (1 - y0) #shape N*1
    X_alpha = alpha * X_tilde_train #broadcasting, shape N*D
    S_N_inv = S_0_inv + np.dot(X_alpha.T, X_tilde_train)
    S_N = np.linalg.inv(S_N_inv)
    return wmap, S_N
    
def bayesian_prediction(X, y, wmap, S_N):
    N_example = X.shape[0]; D_feature = X.shape[1]
    
    mean_a = np.dot(X, wmap).reshape((N_example, 1)) #shape N*1
    variance_a = np.dot(X, np.dot(S_N, X.T)) #shape N*N
    variance_a = np.diag(variance_a).reshape((N_example, 1)) #shape N*1
    kappa_a = np.reciprocal(np.sqrt(1 + np.pi * variance_a / 8)) #shape N*1
    prediction = logistic(kappa_a * mean_a) #shape N*1

    #compute model evidence, i.e. normalisation constant
    ln_PD = log_likelihood(X, y, wmap) - 0.5*D_feature*np.log(N_example)
    PD = np.exp(ln_PD)
    
    return prediction, PD
```

### Confusion matrix

```python
# wmap
def obtain_confusion_matrix(X_tilde_test, w, y_test):
    prediction = predict(X_tilde_test, w)
    #print(prediction[:10])
    
    prediction_hard = np.zeros(prediction.shape)
    for index, element in enumerate(prediction):
        if element > 0.5:
            prediction_hard[index] = 1
            
    confusion_matrix = np.zeros((2, 2))
    for x_predict, y in zip(prediction_hard, y_test):
        if x_predict==0 and y==0:
            confusion_matrix[0][0] += 1
        elif x_predict==1 and y==0:
            confusion_matrix[0][1] += 1
        elif x_predict==0 and y==1:
            confusion_matrix[1][0] += 1
        elif x_predict==1 and y==1:
            confusion_matrix[1][1] += 1
    confusion_matrix /= np.sum(confusion_matrix,1).reshape(2, 1)
    
    return confusion_matrix
obtain_confusion_matrix(X_tilde_test, wmap, y_test)
```

```python
def grid_search(X_train, y_train, X_test, y_test):
    a = np.linspace(-2, 0, 10); l_list = [10**x for x in a]
    b = np.linspace(-1, 1, 10); var_list = [10**y for y in b]
    max_f_val = -10000; max_ll_train = -10000; max_ll_test = -10000
    l_val = 0; var_val = 0
    PD_list_train = []; PD_list_test = []
    best_conf = None; best_wmap = None; best_S_N = None
    
    for i in range(10):
        for j in range(10):
            l = l_list[i]; variance = var_list[j]
            X_tilde_train = get_x_tilde(evaluate_basis_functions(l, X_train, X_train))
            X_tilde_test = get_x_tilde(evaluate_basis_functions(l, X_test, X_train))
            
            wmap, S_N = bayesian_training(X_tilde_train, y_train, X_tilde_test, y_test, variance)
            log_likelihood_train, ln_PD_train = log_likelihood_bay_2(X_tilde_train, y_train, wmap, S_N)
            log_likelihood_test, ln_PD_test = log_likelihood_bay_2(X_tilde_test, y_test, wmap, S_N)
            PD_list_train.append(ln_PD_train); PD_list_test.append(ln_PD_test)
            #evalute models using f1 score
            conf_m = obtain_confusion_matrix_bay(X_tilde_test, wmap, y_test, S_N)
            f1_score = 2*conf_m[1][1] / (2*conf_m[1][1] + conf_m[1][0] + conf_m[0][1])

            if f1_score > max_f_val:
                max_f_val = f1_score
                max_ll_train = log_likelihood_train
                max_ll_test = log_likelihood_test
                l_val = l; var_val = variance
                best_conf = conf_m
                best_wmap = wmap
                best_S_N = S_N
    return max_ll_train, max_ll_test, l_val, var_val, PD_list_train, PD_list_test, best_conf, best_wmap, best_S_N
```
