```python
import torch as t
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t.manual_seed(100)
dtype = t.float
```

```python
x = t.unsqueeze(t.linspace(-1, 1, 50), dim = 1)
y = 3 * x.pow(2) + 2 + t.rand(x.size()) * 0.4
```

```python
plt.scatter(x.numpy(), y.numpy())
plt.show()
```

```python
w = t.randn(1, 1, dtype=dtype, requires_grad=True)
b = t.zeros(1, 1, dtype=dtype, requires_grad=True)
```

```python
lr = 0.001

for i in range(1200):
    y_pred = x.pow(2).mm(w) + b
    loss = 0.5 * (y_pred - y) ** 2
    loss = loss.sum()
    
    loss.backward()
    
    with t.no_grad():
        w -= lr * w.grad + 0.001 * w
        b -= lr * b.grad + 0.001 * w
        
        w.grad.zero_()
        b.grad.zero_()
```

```python
plt.plot(x.numpy(), y_pred.detach().numpy())
plt.scatter(x.numpy(), y.numpy())
plt.show()
```
